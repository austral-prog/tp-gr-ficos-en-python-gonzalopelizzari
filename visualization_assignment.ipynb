```python
# visualization_assignment.ipynb

# -----------------------------
# Importar librerías necesarias
# -----------------------------
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import os

# Crear carpeta para guardar imágenes
os.makedirs("imgs", exist_ok=True)

# -----------------------------
```

```python
# 1. Gráfico de línea - Pobreza en Argentina (1990-2020)
# -----------------------------
"""
### Gráfico de línea: Porcentaje de pobreza en Argentina (1990-2020)
Este gráfico de líneas muestra cómo ha variado el porcentaje de pobreza en Argentina cada 5 años entre 1990 y 2020. 
El gráfico permite visualizar claramente las subidas y bajadas en distintos períodos. Es útil para representar evolución a lo largo del tiempo.
"""

years = np.arange(1990, 2021, 5)
pobreza = [33, 28, 45, 54, 38, 30, 35]

plt.figure(figsize=(8, 5))
plt.plot(years, pobreza, marker='o', linestyle='--', color='darkred', label='Pobreza (%)')
plt.title('Porcentaje de pobreza en Argentina (1990-2020)')
plt.xlabel('Año')
plt.ylabel('Porcentaje de pobreza (%)')
plt.grid(True)
plt.legend()
plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
```

```python
# 2. Gráfico de barras - Copas del Mundo ganadas por países
# -----------------------------
"""
### Gráfico de barras: Copas del Mundo ganadas por país
Este gráfico compara la cantidad de Copas del Mundo ganadas por 5 países. Las barras permiten una comparación clara entre ellos. 
Brasil lidera con 5 títulos, seguido por Alemania e Italia con 4, Argentina con 3 y Francia con 2.
"""

paises = ['Brasil', 'Alemania', 'Italia', 'Argentina', 'Francia']
titulos = [5, 4, 4, 3, 2]

plt.figure(figsize=(8, 5))
plt.bar(paises, titulos, color=['green', 'black', 'blue', 'skyblue', 'navy'])
plt.title('Copas del Mundo ganadas por país')
plt.xlabel('País')
plt.ylabel('Cantidad de títulos')
plt.savefig('imgs/bar_chart.png', dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
```

```python
# 3. Histograma - Distribución de edades en un equipo deportivo
# -----------------------------
"""
### Histograma: Distribución de edades en un equipo
Este histograma muestra cómo se distribuyen las edades en un equipo deportivo ficticio. 
Se observa una tendencia centrada en los 25 años, lo cual es común en deportes de alto rendimiento.
"""

edades = np.random.normal(loc=25, scale=4, size=100)

plt.figure(figsize=(8, 5))
plt.hist(edades, bins=10, color='orange', edgecolor='black')
plt.title('Distribución de edades en un equipo')
plt.xlabel('Edad (años)')
plt.ylabel('Frecuencia')
plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
```

```python
# 4. Gráfico de dispersión - Altura vs Peso
# -----------------------------
"""
### Gráfico de dispersión: Relación entre altura y peso
Este scatter plot muestra la relación entre altura y peso en una muestra aleatoria. 
Se puede observar una correlación positiva: a mayor altura, mayor peso. Es útil para analizar relaciones entre dos variables numéricas.
"""

np.random.seed(0)
altura = np.random.normal(170, 10, 100)
peso = altura * 0.5 + np.random.normal(0, 5, 100)

plt.figure(figsize=(8, 5))
plt.scatter(altura, peso, color='purple', alpha=0.7, label='Individuos')
plt.title('Relación entre altura y peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.grid(True)
plt.savefig('imgs/scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------
```

```python
# 5. Gráfico de pastel - Reparto de tiempo en un día
# -----------------------------
"""
### Gráfico de pastel: Distribución del tiempo en un día típico
Este gráfico circular muestra cómo se reparte el tiempo en un día promedio entre distintas actividades. 
Destaca la importancia del sueño y el trabajo en la rutina diaria.
"""

etiquetas = ['Dormir', 'Estudiar', 'Trabajar', 'Ocio', 'Comer']
tiempo = [8, 4, 8, 3, 1]
colores = sns.color_palette('pastel')[0:5]

plt.figure(figsize=(7, 7))
plt.pie(tiempo, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)
plt.title('Distribución del tiempo en un día típico')
plt.savefig('imgs/custom_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
